```python
import pandas as pd
df= pd.read_csv('/Users/elenayun/Desktop/Others/iris.csv')
```

# Create the Decision Tree classifier and visualize it graphically. 

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing data set
train_data, test_data = train_test_split(df, test_size=0.2, random_state=20)

feature_columns=['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

target_column=['Species']

X_train, y_train, X_test, y_test = train_data[feature_columns], train_data[target_column], test_data[feature_columns], test_data[target_column]

```

```python
#Decision tree classifier

from sklearn.tree import DecisionTreeClassifier
treemodel = DecisionTreeClassifier(random_state=200)
treemodel.fit(X_train,y_train)
predictions = treemodel.predict(X_test)
```

```python
# Decision tree visulisations 
import matplotlib.pyplot as plt
from sklearn import tree


fig = plt.figure(figsize=(15,8))
tree.plot_tree(treemodel,
              feature_names = X_train.columns, 
             class_names=y_train['Species'].unique(),
              filled=True)
```

```python
# Get prediction accuracy  
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
