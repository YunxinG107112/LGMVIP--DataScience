```python
import pandas as pd
members = pd.read_csv('/Users/elenayun/Desktop/Others/members.csv')
extra = pd.read_csv('/Users/elenayun/Desktop/Others/song_extra_info.csv')

songs = pd.read_csv('/Users/elenayun/Desktop/Others/songs.csv')



extra= extra.dropna()
songs=songs.dropna()
members=members.dropna()
```

```python
members['registration_init_time'] = pd.to_datetime(members['registration_init_time'], format='%Y%m%d')
members['expiration_date'] = pd.to_datetime(members['expiration_date'], format='%Y%m%d')

# Extract day, year, and month from registration_init_time
members['reg_day'] = members['registration_init_time'].dt.day
members['reg_month'] = members['registration_init_time'].dt.month
members['reg_year'] = members['registration_init_time'].dt.year

# Extract day, year, and month from expiration_date
members['exp_day'] = members['expiration_date'].dt.day
members['exp_month'] = members['expiration_date'].dt.month
members['exp_year'] = members['expiration_date'].dt.year
```

```python
# Merge dataframes

merged= pd.merge(extra,songs,on='song_id')

merged
```

```python
from sklearn.model_selection import train_test_split

train= pd.read_csv('/Users/elenayun/Desktop/Others/train.csv')

test= pd.read_csv('/Users/elenayun/Desktop/Others/test.csv')

train=train.dropna()
```

```python
train_members=pd.merge(train,members)

train_all=pd.merge(train_members,merged)
```

```python
for col in train_all.select_dtypes(include=['object']).columns:
    train_all[col] = train_all[col].astype('category')
    
for col in train_all.select_dtypes(include=['category']).columns:
    train_all[col] = train_all[col].cat.codes
```

```python
X=train_all[['artist_name',
             'song_length','lyricist','composer','reg_day','exp_day','reg_month',
            'exp_month']]
y=train_all['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report
# Create XGB Classifier 
xgb_model = xgb.XGBClassifier(n_jobs=5,max_depth=10, n_estimators=500)

# Fit the model to training data
xgb_model.fit(X_train, y_train)

# Make predictions on test data
y_pred = xgb_model.predict(X_test)

# Produce classification matrix
print(classification_report(y_test,y_pred))
# Confusion matrix
pd.crosstab(y_test, y_pred, rownames=["Actual"], colnames=["Predicted"])
```

```python
importance = pd.Series(xgb_model.feature_importances_,index=X.columns)
importance.nlargest(5).plot(kind='barh')
```

```python
#random forest model (rf) (initial version)
from sklearn.ensemble import RandomForestClassifier

# Random forest model
clf_rf = RandomForestClassifier(max_depth=2, random_state=0).fit(X_train, y_train)

# Produce true vs predicted label 
predictions_rf = clf_rf.predict(X_test)
# Produce classification matrix
print(classification_report(y_test,predictions_rf))
# Confusion matrix
pd.crosstab(y_test, predictions_rf, rownames=["Actual"], colnames=["Predicted"])
```
